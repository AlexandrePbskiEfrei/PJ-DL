```
def daily_rate(stock_name, start, stop):
    cursor = cnxn.cursor()
    cursor.execute("SELECT [company_name],[Date], [Close] FROM [dbo].[Stocks] WHERE [company_name] = '"+ stock_name +"' AND [Date] BETWEEN '" + start + "' AND '" + stop +"'")
    row = cursor.fetchone()
    while row:
        print (str(row[0]) + " " + str(row[1])+ " " + str(row[2]))
        row = cursor.fetchone()
print("USING SQL\n")
daily_rate("FACEBOOK", "2017-01-03", "2020-01-01")

def daily_rate_df(stock_name, start, stop):
    mask = (df["Date"] >= start) & (df["Date"] <= stop) & (df["company_name"] == stock_name)
    return df[mask]
print("\nUSING DATAFRAME\n")
df_daily = daily_rate_df("FACEBOOK","2017-01-03", "2020-01-01")
df_daily.head()
```

```
def moving_average(stock_name, start, stop, nb_mp):
    mask = (df["Date"] >= start) & (df["Date"] <= stop) & (df["company_name"] == stock_name)
    df_r = df[mask]
    mvg_avg = df_r["Open"].rolling(nb_mp).mean()
    return mvg_avg

df_mvg_avg = moving_average("FACEBOOK", "2017-01-03", "2020-01-01", 5)
df_mvg_avg
```
